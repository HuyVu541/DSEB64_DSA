```python
#add left, add right, is root, is leaf, check ancestor, height, depth, print
from collections import deque


        

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        self.parent = None

    def add_left(self, branch):
        self.left = branch
        branch.parent = self

    def add_right(self, branch):
        self.right = branch
        branch.parent = self

    def is_root(self):
        return not self.parent

    def is_leaf(self):
        return not (self.left or self.right)

    def is_ancestor(self, ancestor):
        while self is not None:
            if self is ancestor:
                return True
            self = self.parent
        return False
    
    def height(self, node): # Input root of tree
        if node is None:
            return 0
        return 1 + max(self.height(node.left), self.height(node.right))
    
    def depth(self):
        count = 0
        while self.parent is not None:
            self = self.parent
            count +=1
        return count
    
    def list_nodes_DFS(self):
        print(self.data)
        if self.left:
            self.left.list_nodes()
        if self.right:
            self.right.list_nodes()
    
    def sum_right(self, node):
        total  = 0  
        if node:
            if (node.is_leaf()) and (node.parent.right is node):
                return node.data
            total += self.sum_right(node.left)
            total += self.sum_right(node.right)
        return total

    def preorder_traversal_helper(self):
        print(self.data)
        
        if self.left:
            self.left.preorder_traversal_helper()
        if self.right:
            self.right.preorder_traversal_helper()


    def sum_right2(self, node):
        total = 0
        if node:
            if node.right:
                if node.right.is_leaf():
                    total += node.right.data
                else:
                    total += self.sum_right(node.right)
            if node.left:
                total += self.sum_right(node.left)
        return total

    def draw_tree(self):
        curr = self.height(self)*2 - 1
        res = []
        

    def inorder_traversal(self):
        if self.left:
            self.left.inorder_traversal()
        print(self.data)
        if self.right:
            self.right.inorder_traversal()        

    def list_nodes_by_level(self):
        level_count = 0
        arr = deque()
        arr.append(self)
        while len(arr) > 0:
            print(f"Level {level_count}: ")
            for i in range(len(arr)):
                node = arr.popleft()
                if node:
                    print(node.data, end = " ")
                    if node.left:
                        arr.append(node.left)
                    if node.right:
                        arr.append(node.right)
            print("\n")
            level_count += 1

    def list_nodes_by_level2(self):
        level_count = 0
        arr = deque()
        arr.append(self)
        arr.append(None)
        while len(arr) > 0:
            node = arr.popleft()
            if node:
                print(node.data, end = " ")
                if node.left:
                    arr.append(node.left)
                if node.right:
                    arr.append(node.right)
            else:
                if len(arr) > 0:
                    arr.append(None)
                print("\n")
class BinaryTree():
    def __init__(self, root):
        self.root = root
    def preorder_traversal(self):
        self.root.preorder_traversal_helper()
    def print_tree(self):
        tree_height = self.root.height(self.root)
        curr = tree_height**2-1
        res = [[" " for i in range(curr)] for i in range(tree_height)]
        arr = deque()
        arr.append(self.root)
        level_count = 0
        while len(arr) > 0:
            count = 1
            end = True
            for i in range(len(arr)):  
                node = arr.popleft()
                if node:
                    end = False
                    res[level_count][(curr//2+1)*count-1] = str(node.data)
                    arr.append(node.left)
                    arr.append(node.right)
                else:
                    arr.append(None)
                    arr.append(None)
                count +=2
            level_count +=1
            curr = curr//2
            if end: #All values are None
                break
        res = ["".join(row) for row in res]
        for row in res:
            print(row)
        
```
